### Aggregating values and number of products per order

Prices of the ordered products are in order_item datset. Let's calculate sum and mean values of products, and number of products per order, and then add it to orders dataset:

```python
order_items_df['total_value']=order_items_df['price']+order_items_df['freight_value']
```

```python
order_agg1=order_items_df[['order_id','price','freight_value','total_value']].groupby('order_id').sum().reset_index()
```

```python
order_agg1.columns=['order_id','sum_price','sum_freight_value','sum_total_value']
```

```python
order_agg2=order_items_df[['order_id','price','freight_value','total_value']].groupby('order_id').mean().reset_index()
```

```python
order_agg2.columns=['order_id','avg_price','avg_freight_value','avg_total_value']
```

```python
order_df=order_df.merge(order_agg1,on='order_id')
```

```python
order_df=order_df.merge(order_agg2,on='order_id')
```

```python
order_df['num_items']=order_df['sum_total_value']/order_df['avg_total_value']
```

```python
list(order_df.columns)
```

### Aggregating most-common product categories per order

We would like to aggregate information about product categories per order. We can do it by creating binary features corresponding to the most common product names and aggregate those features per orders:

```python
most_common_categ=items_df['product_category_name'].value_counts().reset_index()[0:10]
```

```python
most_common_categ
```

```python
most_common_categ_names=list(most_common_categ['index'])
```

```python
most_common_categ_names
```

```python
produc_category_bin=pd.get_dummies(items_df['product_category_name'])
```

```python
most_common_bin=produc_category_bin[most_common_categ_names]
```

```python
items_df.columns
```

```python
most_common_bin['product_id']=items_df['product_id']
```

```python
order_items_df=order_items_df.merge(most_common_bin,on='product_id')
```

```python
order_items_df.columns
```

```python
order_items_df_agg=order_items_df[['order_id']+most_common_categ_names].groupby('order_id').sum().reset_index()
```

```python
order_items_df_agg
```

```python
order_df=order_df.merge(order_items_df_agg,on='order_id')
```

```python
list(order_df.columns)
```
